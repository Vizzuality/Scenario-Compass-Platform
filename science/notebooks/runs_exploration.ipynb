## Exploring the Runs Endpoint

```python
# Basic statistics
print("Basic statistics:")
runs_df.describe(include='all')
```

## Analyzing Run Characteristics

```python
# Analyze models
if 'model__id' in runs_df.columns:
    print("Models in the dataset:")
    model_counts = runs_df['model'].value_counts()
    print(model_counts)

    # Visualize model distribution
    plt.figure(figsize=(12, 6))
    model_counts.plot(kind='bar')
    plt.title('Number of Runs by Model')
    plt.xlabel('Model')
    plt.ylabel('Number of Runs')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
# Analyze scenarios
if 'scenario__id' in runs_df.columns:
    print(f"\nTotal number of scenarios: {runs_df['scenario'].nunique()}")

    # Show top 20 scenarios
    scenario_counts = runs_df['scenario'].value_counts().head(20)
    print("\nTop 20 scenarios by run count:")
    print(scenario_counts)

    # Visualize top scenarios
    plt.figure(figsize=(12, 8))
    scenario_counts.plot(kind='barh')
    plt.title('Top 20 Scenarios by Number of Runs')
    plt.xlabel('Number of Runs')
    plt.ylabel('Scenario')
    plt.tight_layout()
    plt.show()
```
